<h3>Target Guided Ordinal Encoding</h3>
It is a technique used to encode categirical variables based on their relartionshipwith target variable. This encoding is useful when we have categorical variables with a large number of unique categories<br>
In target guided ordinal encoding, we replace each category in thr categorical variable with the numerical value based on the mean or median of the target variable for that category.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'city': ['New York', 'Mumbai', 'New Jersey', 'Delhi','Paris','New York','Mumbai','Delhi'],
    'Rent' :[400,200,300,150,300,250,600,450]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['Rent'].mean()
```

```python
mean_price
```

```python
df['city encoded'] = df['city'].map(mean_price)
```
