````python
import pandas as pd
import numpy as np
````

 Python list <b>headers</b> containing name of headers


````python
# These are the headers which you need to link to your dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
````

````python
df = pd.read_csv('car_dataset.data', names = headers)
````

````python
# To see what the data set looks like, we'll use the head() method.
df.head()
````

````python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
````

````python
missing_data = df.isnull()
missing_data.head(5)
````

"True" stands for missing value, while "False" stands for not missing value.


````python
for column in headers:
    print(column)
    print (missing_data[column].value_counts())
    print("")
````

<h2 id="deal_missing_values">Deal with missing data</h2>
<br>
<h3>How to deal with missing data?</h3>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
    </li>
</ol>


Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four".
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur, also adding mean in no. of doors would be illogical</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
        </ul>
    </li>
</ul>


Replacing Nan with mean in normalized losses columns


````python
avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
````

````python
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_norm_loss)
df
````

Replacing Nan with mean in bore column


````python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
````

````python
df["bore"] = df['bore'].replace(np.nan, avg_bore)
df
````

Replacing Nan with mean in stroke

````python
avg_bore=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_bore)
````

````python
df['stroke'] = df["stroke"].replace(np.nan, avg_bore)
df
````

Replacing Nan with mean in horsepower column


````python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
````

````python
df['horsepower'] = df['horsepower'].replace(np.nan, avg_horsepower)
df
````

Replacing Nan with mean in peak-rpm column

````python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
````

````python
df['peak-rpm'] = df['peak-rpm'].replace(np.nan, avg_peakrpm)
df
````

Replacing Nan with most frequent value in num of doors

````python
df['num-of-doors'].value_counts()
````

We can see that four doors are the most common type. We can also use the ".idxmax()" method to calculate for us the most common type automatically


````python
df['num-of-doors'].value_counts().idxmax()
````

````python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"] = df["num-of-doors"].replace(np.nan, "four")
df
````

Finally, let's drop all rows that do not have price data


````python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
````

````python
new_missing_data = df.isnull()

for column in headers:
    print (new_missing_data[column].value_counts())
    print("")
````
