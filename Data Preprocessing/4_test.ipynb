```python
mean_rating = data['RATING'].astype('float').mean(axis = 0)
mean_rating
```

```python
data['RATING'] = data['RATING'].replace(np.NAN,mean_rating)
```

```python
data['RATING'].isnull().sum()
```

```python
mode_votes = data['VOTES'].value_counts()
mode_votes
```

```python
mode_votes = data['VOTES'].value_counts().idxmax()
mode_votes
```

```python
data['VOTES'] = data['VOTES'].replace(np.NAN, 7)
```

```python
data['VOTES'].isnull().sum()
```

```python
mean_runtime = data['RunTime'].astype('float').mean(axis = 0)
mean_runtime
```

```python
data['RunTime'] = data['RunTime'].replace(np.NAN,mean_runtime)
```

```python
data['RunTime'].isnull().sum()
```

```python
data['Gross'].value_counts()
```

```python
data['Gross'].isnull().value_counts()
```

```python
data['Gross'] = data['Gross'].str.replace('$', '').str.replace('M','').str.replace(',','').astype('float')
```

```python
mean_gross = data['Gross'].astype('float').mean(axis = 0)
mean_gross
```

```python
data['Gross'] = data['Gross'].replace(np.NAN,mean_gross)
```

```python
data['Gross'].isnull().sum()
```
