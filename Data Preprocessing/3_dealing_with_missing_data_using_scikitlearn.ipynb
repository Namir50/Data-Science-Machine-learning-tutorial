```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
imputer_mean  = SimpleImputer(missing_values=np.nan,strategy = 'mean')
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
data = pd.read_csv('car_dataset.data',names = headers_data)
data = pd.DataFrame(data)
data
```

```python
data = data.replace('?',np.nan)
```

```python
data
```

```python
data.info()
```

```python
missing_data = data.isnull()
missing_data
```

```python
for columns in headers_data:
    print(missing_data[columns].value_counts())
    print()
```

The columns that require mean for Nan are mentioned below

```python
data[['normalized-losses','stroke','bore','horsepower','peak-rpm']] = imputer_mean.fit_transform(
   data[['normalized-losses','stroke','bore','horsepower','peak-rpm']]).astype('object')
data
```

```python
data.head(5)
```

Columns that require mode in Nan are mentioned before

```python
imputer_mode = SimpleImputer(missing_values=np.nan,strategy= 'most_frequent')
```

```python
data[['num-of-doors']] = imputer_mode.fit_transform(data[['num-of-doors']]).astype('object')
```

```python
data.info()
```

Price column needs where Nan row should be dropped

```python
data.dropna(subset = ['price'],axis = 0,inplace=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.info()
```
