```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,2,31,44,3]
```

Creating a line plot

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Basic line plot")
plt.show
```

```python
plt.plot(x,y,color = 'red',linestyle = '-')
```

```python
plt.plot(x,y,color = 'red',linestyle = '--')
```

```python
plt.plot(x,y,color = 'red',linestyle = '-.')
```

```python
plt.plot(x,y,color = 'blue',linestyle = '--',marker = 'o')
```

```python
plt.plot(x,y,color = 'blue',linestyle = '--',marker = '*',markersize = 9, linewidth = 3)
```

Multiplots

```python
x = [1,2,3,4,5]
y1 = [10,23,43,91,5]
y2 = [96,33,12,43,21]
```

```python
plt.plot(x,y1,y2)
```
