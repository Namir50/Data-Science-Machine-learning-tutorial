```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [10,2,31,44,3]
```

Creating a line plot

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Basic line plot")
plt.show
```

```python
plt.plot(x,y,color = 'red',linestyle = '-')
```

```python
plt.plot(x,y,color = 'red',linestyle = '--')
```

```python
plt.plot(x,y,color = 'red',linestyle = '-.')
```

```python
plt.plot(x,y,color = 'blue',linestyle = '--',marker = 'o')
```

```python
plt.plot(x,y,color = 'blue',linestyle = '--',marker = '*',markersize = 9, linewidth = 3)
```

Multiplots

```python
x = [1,2,3,4,5]
y1 = [10,23,43,91,5]
y2 = [96,33,12,43,21]

plt.figure(figsize = (9,5))
plt.subplot(1,2,1)  #1 row 2 column
plt.plot(x,y1,color = 'green')
plt.title('Plot 1')

plt.subplot(1,2,2)  #1 row 2 column
plt.plot(y1,x,color = 'green')
plt.title('Plot 2')

plt.subplot(2,2,3)  #1 row 2 column
plt.plot(x,y2,color = 'green')
plt.title('Plot 3')
```

Pie chart

```python
nums = [10,40,30,20]
categories = ['A','B','C','D']
colors = ['gold','blue','red','orange']

plt.pie(nums,labels=categories, colors=colors,autopct="%1.1f%%",shadow = True)
```

```python
nums = [10,40,30,20]
categories = ['A','B','C','D']
colors = ['gold','blue','red','orange']
explode = (0.1,0,0,0) #moves out defined slice

plt.pie(nums,labels=categories,explode = explode, colors=colors,autopct="%1.1f%%",shadow = True)
```
