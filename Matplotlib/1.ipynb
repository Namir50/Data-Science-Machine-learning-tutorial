```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,2,31,44,3]
```

Creating a line plot

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Basic line plot")
plt.show
```

```python
plt.plot(x,y,color = 'red',linestyle = '-')
```

```python
plt.plot(x,y,color = 'red',linestyle = '--')
```

```python
plt.plot(x,y,color = 'red',linestyle = '-.')
```
