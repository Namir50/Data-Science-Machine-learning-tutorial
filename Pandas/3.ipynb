```python
import pandas as pd
```

```python
data = pd.read_csv("Employee.csv")
```

```python
data
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age']).mean().astype(int)
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mean()).astype(int)
```

```python
data.isnull().sum()
```

```python
data['Department'] = data['Department'].fillna(data['Department'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data
```

```python
data['Salary with Bonus'] = data['Salary'] * 2
```

```python
data
```

```python
data.describe
```

```python
data.isnull().sum()
```

```python
data = data.reindex(columns=['ID','Name','Age','Gender','Department','Salary','Salary with Bonus'])
```

```python
data
```

```python
data.groupby('Department')['Salary'].mean().astype(int)
```

```python
data.groupby('Gender')['Salary'].mean()
```

```python
data.groupby('Department')['Salary'].agg(['mean','std','count','sum'])
```
