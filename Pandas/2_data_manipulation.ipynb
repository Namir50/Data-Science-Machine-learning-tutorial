Data Aggregation

```python
data["Salary with Bonus"] = data['Salary Complete'].apply(lambda x: x *2)
```

```python
data
```

```python
grouped_mean = data.groupby("City")['Salary with Bonus'].mean().astype(int) #Using "Int64" instead of int is applicable when you're dealing with pandas Nullable Integer type, which supports missing values (NaN). If your Salary with Bonus column might contain missing values, using .astype("Int64") ensures compatibility with those null entries.
print(grouped_mean)
```

```python
grouped_sum = data.groupby("Country")['Salary with Bonus'].sum()
print(grouped_sum)
```

```python
grouped_agg = data.groupby('Country')['Salary with Bonus'].agg(['mean','sum','count'])
grouped_agg
```
